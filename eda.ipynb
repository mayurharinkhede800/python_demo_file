```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
visa_df=pd.read_csv(r'C:\Users\HP\Documents\NareshIT\Data_Files\Visadataset.csv')
```

```python
categorical=visa_df.select_dtypes(include="O").columns
numerical=visa_df.select_dtypes(exclude="O").columns
```

```python
keys=visa_df['continent'].value_counts().keys()
values=visa_df['continent'].value_counts().values
cont_df=pd.DataFrame(zip(keys,values))
cont_df
```

```python
uniq=visa_df['continent'].unique()
for i in uniq:
    print(f'{i}:{len(visa_df[visa_df['continent']==i])}')
```

```python
unique=visa_df['continent'].unique()
counts=[]
for i in unique:
    con=visa_df['continent']==i
    count=len(visa_df[con])
    counts.append(count)
pd.DataFrame(zip(unique,counts),columns=['continent','values'],index=[1,2,3,4,5,6])
```

```python
visa_df['continent'].value_counts()
```

```python
keys=visa_df['continent'].value_counts().keys()
cols=visa_df['continent'].value_counts().values
pd.DataFrame(zip(keys,cols),columns=['continent','values'],index=[1,2,3,4,5,6])
```

```python
visa_df.groupby('continent').size()
```

```python
plt.figure(figsize=(10,5))
# col=['red','green','yellow','blue']
plt.bar(keys,cols)
plt.title('continent Data')
plt.xlabel('continent')
plt.ylabel('values')
plt.savefig('continent1.jpg')
plt.show()
```
